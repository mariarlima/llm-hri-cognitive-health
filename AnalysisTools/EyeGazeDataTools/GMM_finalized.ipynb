```python
import os
import re
import numpy as np
from GMM_analysis import get_eye_gaze_data, get_GMM_baseline, get_normalized_log_likelihood, get_Overall_Mask_GMM_baseline

directory_path = "./data/"
output_filepath = "./data_processed/Image_Components_GMM_results.csv"
output_filepath_overall_gmm = "./data_processed/Overall_GMM_results.csv"
output_filepath_overfitted_overall_gmm = "./data_processed/Overfitted_Overall_GMM_results.csv"

verbose = True

pattern = r'(P\d{2}_S\d)\s?(#\s?)?.*?_all_gaze\.csv'
# Get all file names in the directory
file_names = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
baseline_cookie = get_GMM_baseline("Cognitive Picture Description Task", verbose=verbose)
overall_baseline_cookie = get_Overall_Mask_GMM_baseline("Cognitive Picture Description Task", verbose=verbose)
baseline_picnic = get_GMM_baseline("Picnic Task", verbose=verbose)
overall_baseline_picnic = get_Overall_Mask_GMM_baseline("Picnic Task", verbose=verbose)
overfitted_overall_baseline_cookie = get_Overall_Mask_GMM_baseline("Cognitive Picture Description Task", verbose=verbose, num_components=50)
overfitted_overall_baseline_picnic = get_Overall_Mask_GMM_baseline("Picnic Task", verbose=verbose, num_components=50)

print(baseline_cookie)
print(overall_baseline_cookie)
print(baseline_picnic)
print(overall_baseline_picnic)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def get_normalized_log_likelihood(points, gmm, x_scale=1920, y_scale=1080, show_plot=True):
    log_llh = gmm.score(points)
    # log_llh /= points.shape[0]
    # TODO: Add titles Pxx Sx
    show_plot = True
    if show_plot:
        df_describe = pd.DataFrame(points)
        print(df_describe.describe())
        plt.scatter(points[:, 0], points[:, 1], marker='x', s=1)
        plt.show()
    return log_llh
    
# Print all file names
for file_name in file_names:
    match = re.search(pattern, file_name)
    if match:
        # Extracting the desired part
        extracted_string = match.group(1).replace('_', ' ')
        
        print(f"Processing: {extracted_string}")
        # print(f"File name: {file_name}")
        try:
            points, task = get_eye_gaze_data(f"{directory_path}{file_name}")
            points[:, 0] *= 1920
            points[:, 1] *= 1080
            baseline_gmm = baseline_cookie if task == "Cognitive Picture Description Task" else baseline_picnic
            overall_baseline_gmm = overall_baseline_cookie if task == "Cognitive Picture Description Task" else overall_baseline_picnic
            overfitted_baseline_gmm = overfitted_overall_baseline_cookie if task == "Cognitive Picture Description Task" else overfitted_overall_baseline_picnic
            score = get_normalized_log_likelihood(points, gmm=baseline_gmm, show_plot=False)
            score_overall_baseline = get_normalized_log_likelihood(points, gmm=overall_baseline_gmm, show_plot=False)
            score_overfitted_baseline = get_normalized_log_likelihood(points, gmm=overfitted_baseline_gmm, show_plot=False)
        except ValueError:
            score = float('nan')
            score_overall_baseline = float('nan')
            score_overfitted_baseline = float('nan')
        print(f"{extracted_string} GMM score: {score:.3f}")
        print(f"{extracted_string} Overall GMM score: {score_overall_baseline:.3f}")
        print(f"{extracted_string} Overfitted Overall GMM score: {score_overfitted_baseline:.3f}")
        with open(output_filepath, "a") as f:
            f.write(f"{extracted_string},{score:.3f}\n")
        with open(output_filepath_overall_gmm, "a") as f_o:
            f_o.write(f"{extracted_string},{score_overall_baseline:.3f}\n")
        with open(output_filepath_overfitted_overall_gmm, "a") as f_o_o:
            f_o_o.write(f"{extracted_string},{score_overfitted_baseline:.3f}\n")
```

```python
def get_mean_entropy(points, gmm):
    probs = gmm.predict_proba(points)
    entropies = -np.sum(probs * np.log(probs + 1e-10), axis=1)  # add epsilon to avoid log(0)
    mean_entropy = np.mean(entropies)
    show_plot = True
    if show_plot:
        df_describe = pd.DataFrame(points)
        print(df_describe.describe())
        plt.scatter(points[:, 0], points[:, 1], marker='x', s=1)
        plt.show()
    return mean_entropy


entropy_output_filepath = "./data_processed/Image_Components_Entropy_results.csv"
entropy_output_filepath_overall_gmm = "./data_processed/Overall_Entropy_results.csv"
entropy_output_filepath_overfitted_overall_gmm = "./data_processed/Overfitted_Entropy_results.csv"

for file_name in file_names:
    match = re.search(pattern, file_name)
    if match:
        # Extracting the desired part
        extracted_string = match.group(1).replace('_', ' ')
        
        print(f"Processing: {extracted_string}")
        # print(f"File name: {file_name}")
        try:
            points, task = get_eye_gaze_data(f"{directory_path}{file_name}")
            points[:, 0] *= 1920
            points[:, 1] *= 1080
            baseline_gmm = baseline_cookie if task == "Cognitive Picture Description Task" else baseline_picnic
            overall_baseline_gmm = overall_baseline_cookie if task == "Cognitive Picture Description Task" else overall_baseline_picnic
            overfitted_baseline_gmm = overfitted_overall_baseline_cookie if task == "Cognitive Picture Description Task" else overfitted_overall_baseline_picnic
            ic_entropy = get_mean_entropy(points, gmm=baseline_gmm) #, show_plot=False)
            overall_entropy = get_mean_entropy(points, gmm=overall_baseline_gmm) #, show_plot=False)
            overfitted_entropy = get_mean_entropy(points, gmm=overfitted_baseline_gmm) #, show_plot=False)
        except ValueError:
            ic_entropy = float('nan')
            overall_entropy = float('nan')
        print(f"{extracted_string} Mean Entropy: {ic_entropy:.3f}")
        print(f"{extracted_string} Overall Mean Entropy: {overall_entropy:.3f}")
        print(f"{extracted_string} Overfitted Overall Mean Entropy: {overfitted_entropy:.3f}")
        with open(entropy_output_filepath, "a") as f:
            f.write(f"{extracted_string},{ic_entropy:.3f}\n")
        with open(entropy_output_filepath_overall_gmm, "a") as f_o:
            f_o.write(f"{extracted_string},{overall_entropy:.3f}\n")
        with open(entropy_output_filepath_overfitted_overall_gmm, "a") as f_o_o:
            f_o_o.write(f"{extracted_string},{overfitted_entropy:.3f}\n")
```
