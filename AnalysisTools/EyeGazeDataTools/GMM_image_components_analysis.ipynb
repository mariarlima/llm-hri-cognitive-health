```python
def get_per_image_component_results(filepath, baseline, labels_map):
    points, task = get_eye_gaze_data(filepath, True)
    eye_gaze_coords = points[:, :2]
    fixations = points[:, 2:]
    eye_gaze_coords[:, 0] *= 1920
    eye_gaze_coords[:, 1] *= 1080
    gmm_component_prediction = baseline.predict(eye_gaze_coords)
    image_component_prediction = np.vectorize(labels_map.get)(gmm_component_prediction)
    analysis_results = {"Total": 0.0}
    analysis_results_percentage = {"Total": 1.0}
    for image_component_label in np.unique(image_component_prediction):
        # print(image_component_label)
        analysis_results[image_component_label] = 0.0
    for image_component_label, fixation_duration in zip(image_component_prediction, fixations):
        analysis_results[image_component_label] += fixation_duration
        analysis_results["Total"] += fixation_duration
    for key in analysis_results:
        analysis_results_percentage[key] = analysis_results[key] / analysis_results["Total"]
    for key in analysis_results_percentage:
        print(f"{key}: {analysis_results_percentage[key][0] * 100:0.2f} %")
        analysis_results_percentage[key] = analysis_results_percentage[key][0]
        analysis_results[key] = analysis_results[key][0]
    return analysis_results, analysis_results_percentage
```

```python
time, percentage = get_per_image_component_results("./data/P02_S1_fixations.csv", baseline_cookie, labels_map_cookie)
```

```python
import os
import numpy as np
import re
from GMM_analysis import get_eye_gaze_data, get_GMM_baseline_with_component_label_map, get_normalized_log_likelihood

directory_path = "./data/"
output_filepath = "./data_processed/GMM_results.csv"

pattern = r'(P\d{2}_S\d)\s?(#\s?)?.*?_all_gaze\.csv'
# Get all file names in the directory
file_names = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
baseline_cookie, lables_map_cookie = get_GMM_baseline_with_component_label_map("Cognitive Picture Description Task", verbose=True)
baseline_picnic, lables_map_picnic = get_GMM_baseline_with_component_label_map("Picnic Task", verbose=False)
```

```python
# other_pid = [5, 6, 7, 8, 9, 10, 12, 13, 14, 19, 22, 24]

# cognitive_normal_pid = [2, 4, 15, 18, 21, 23, 25]

# Remove bad data
# P15 S2, S4 double check
other_pid = ['P6', 'P7', 'P8', 'P9', 'P10', 'P12', 'P13', 'P14', 'P16', 'P19', 'P22', 'P24']

cognitive_normal_pid = ['P2', 'P15', 'P18', 'P21', 'P23', 'P25']

excluded_pid = [1, 3, 11]
```

```python

for file_name in file_names:
    match = re.search(pattern, file_name)
    if match:
        # Extracting the desired part
        extracted_string = match.group(1).replace('_', ' ')
        print(f"Processing: {extracted_string}")
        # print(f"File name: {file_name}")
        baseline_gmm = None
        labels_map = None
        temp = extracted_string.split(" ")
        group = "Excluded"
        if temp[0] in cognitive_normal_pid:
            group = "Cognitive Normal"
        elif temp[0] in other_pid:
            group = "Other"
        try:
            points, task = get_eye_gaze_data(f"{directory_path}{file_name}")
            baseline_gmm = baseline_cookie if task == "Cognitive Picture Description Task" else baseline_picnic
            labels_map = lables_map_cookie if task == "Cognitive Picture Description Task" else lables_map_picnic
            time, percentage = get_per_image_component_results(f"{directory_path}{file_name}", baseline_gmm, labels_map)
            for key in time:
                df = pd.concat([df, pd.DataFrame({'PID': [temp[0]], 'Session': [temp[1]], 'Image Component': [key], 'Duration': [time[key]], 'Percentage': [percentage[key]], 'Group':[group]})], ignore_index=True)
        except ValueError:
            time = None
            percentage = None
            print(f"{extracted_string} does not have valid data.")
            image_components = set(labels_map.values())
            for ic in image_components:
                df = pd.concat([df, pd.DataFrame({'PID': [temp[0]], 'Session': [temp[1]], 'Image Component': [ic], 'Duration': [float('nan')], 'Percentage': [float('nan')], 'Group':[group]})], ignore_index=True)
```
