```python
import matplotlib as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/LinguisticFeaturesUPD.csv")
```

```python
df.describe()
```

```python
df_cookie = df[df["Session"].isin(["S1", "S3", "S5"])]
df_picnic = df[df["Session"].isin(["S2", "S4"])]
df_robot = df[df["Session"].isin(["S2", "S3", "S4"])]
```

```python
import seaborn as sns
from utils import plotting
from statannotations.Annotator import Annotator
```

```python
pairs = [("S2", "S3"), ("S3", "S4"), ("S2", "S4")]

with plotting.paper_theme():
    ax = sns.boxplot(data=df_robot, x="Session", y="Task Performance T1", hue="Session")# , width=.5, gap=0.1)
    sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
    ax.set_title('T1 Performance over session')
    annotator = Annotator(ax, pairs, data=df_robot, x="Session", y="Task Performance T1")
    annotator.configure(test='Mann-Whitney', text_format='star', loc='outside')
    annotator.apply_and_annotate()
```
